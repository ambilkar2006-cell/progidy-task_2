```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Create sample dataset
data = {
    'pulse': [80,82,84,85,86,87,88,89,90,91,
              92,93,94,95,96,97,98,99,100,101,
              102,104,105,106,108,110,115,120,125,130,
              135,140,145,150,95,96,97,98,99,100,
              88,89,90,91,92,93,94,95,96,97,
              98,99,100,101,102,103,104,105,90,91,
              92,93,94,95,96,97,98,99,100,101,
              85,86,87,88,89,90,91,92,93,94,
              95,96,97,98,99,100,101,102,103,104],

    'kind': ['rest','walking','running']*30,

    'diet': ['low fat']*45 + ['no fat']*45
}

df = pd.DataFrame(data)

# Display first rows
print(df.head())

# Dataset information
print(df.info())

# Missing values
print(df.isnull().sum())

# Summary statistics
print(df.describe())
```

```python
plt.figure(figsize=(8,5))

plt.hist(df['pulse'], bins=10)

plt.title("Pulse Rate Distribution")
plt.xlabel("Pulse Rate")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8,5))

df['kind'].value_counts().plot(kind='bar')

plt.title("Exercise Type Distribution")
plt.xlabel("Exercise Type")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(8,5))

df['diet'].value_counts().plot(kind='bar')

plt.title("Diet Distribution")
plt.xlabel("Diet Type")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(8,5))

pd.crosstab(df['diet'], df['kind']).plot(kind='bar')

plt.title("Diet vs Exercise Type")
plt.xlabel("Diet")
plt.ylabel("Count")

plt.show()
```
